#데이터 전처리 과정

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#0. 데이터 불러오기 (코랩에서는 필요한 과정)
from google.colab import files
uploaded = files.upload()
```

```python
#1. 데이터 구성
path = "/content/"
"""경로를 지정한 변수를 통해서 편하게 변수를 지정할 수 있음
쌍따옴표 안에서의 숫자 등은 문자로 인식
"문자"+"문자"=문자문자 로 단순 연결시 사용됨"""


train_csv = pd.read_csv("/content/train.csv", index_col=0)
                                              # 첫 번째 컬럼은 데이터가 아니고 인덱스를 의미
                        # . :점 하나는 루트(뿌리)를 의미
# train_csv에 파일의 데이터를 입력

print(train_csv)
```

```python
test_csv = pd.read_csv(path + "test.csv", index_col=0)
print(test_csv)
```

```python
submission_csv = pd.read_csv(path + "submission.csv", index_col=0)
print(submission_csv)
```

```python
print(train_csv.info()) # 컬럼의 데이터 숫자를 확인
print(train_csv.isna().sum())  # 칼럼 내의 결측치를 확인
```

```python
train_csv = train_csv.dropna() # 데이터 내의 결측치를 삭제
print(train_csv.isna().sum())
print(train_csv.info())
print(train_csv)
```

```python
print(test_csv.isna().sum()) # 테스트 데이터의 각 컬럼의 결측치 확인
```

```python
# 테스트 데이터의 경우에는 결측치를 삭제한다면,
#서브미션에 추가해야하는 데이터를 삭제하는 문제가 발생하기 때문에 문제 발생
# NaN은 공백의 데이터이기 때문에 연산을 할 수 없음
# 따라서 임의의 값을 대입해야 하는데, "평균치"를 대입할 수 있다.

test_csv = test_csv.fillna(test_csv.mean())
                    # .fill = 채우는 함수, na = 결측치
                    # mean() 평균값을 구하는 함수
                    # 평균값을 대입하는 경우 대표성을 위해 컬럼의 평균을 대입해야 한다.
print(test_csv.info())
```

# 모델 구성

```python
x = train_csv.drop(['count'], axis=1)
# axis = 0은 행의 [---]이름의 행 삭제, axis =1은 열의 [---]이름의 열 삭제
print(x) #[1328 rows x 9 columns]
y = train_csv['count']
print(y) #Length: 1328
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.15,
                        random_state=10)
```

# 모델구성

```python
model = Sequential()
model.add(Dense(10, input_dim=9))
model.add(Dense(1))
```

# 컴파일 및 훈련

```python
model.compile(loss = 'mse', optimizer = 'adam')
model.fit(x_train, y_train, epochs = 100, batch_size = 10)
```

#평가 및 예측

```python
loss = model.evaluate(x_test,y_test)
result = model.predict([test_csv])
y_predict = model.predict([x_test])
r2 = r2_score(y_test, y_predict)



print(result)
print(loss)
print(r2)
```

# 데이터의 출력
##submission.csv 만들기  
// count 컬럼 값 채우기

```python
submission_csv['count'] = result
print(submission_csv)
print(submission_csv.shape)
```
