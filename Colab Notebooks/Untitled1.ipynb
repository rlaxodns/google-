```python
떡볶이판매수 = int(input('떡볶이는 몇 그릇 팔았나요? '))
라면판매수 = int(input('라면은 몇 그릇 팔았나요? '))
튀김판매수 = int(input('튀김은 몇 개 팔았나요? '))

떡볶이매출 = 떡볶이판매수 * 2000
라면매출 = 라면판매수 * 3000
튀김매출 = 튀김판매수 * 500
총매출 = 떡볶이매출 + 라면매출 + 튀김매출

print(f'-----<오늘의 매출 정산>-----')
print(f'떡볶이의 매출: {떡볶이매출}원')
print(f'라면의 매출: {라면매출}원')
print(f'튀김의 매출: {튀김매출}원')
print(f'오늘의 총 매출은 {총매출}원입니다.')
```
